```python
import MeCab

text = 'これは日本語の文章です．'

tagger1 = MeCab.Tagger('-Owakati')
print(tagger1.parse(text).split())

tagger2 = MeCab.Tagger()
print(tagger2.parse(text))
```

```python
from janome.tokenizer import Tokenizer


t1 = Tokenizer(wakati=True)
print(t1.tokenize(text))

#t2 = Tokenizer('userdict.csv')
t2 = Tokenizer()
for token in t2.tokenize(text):
    print(token)
```
