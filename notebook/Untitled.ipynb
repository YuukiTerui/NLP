# テキストの入手

```python
import urllib.request
import zipfile

import re


def get_text(url, zip_):
    urllib.request.urlretrieve(url, zip_)
    with zipfile.ZipFile(zip_, 'r') as myzip:
        myzip.extractall()
        for myfile in myzip.infolist():
            filename = myfile.filename
            print(filename)
            with open(filename, encoding='sjis') as file:
                test_org = file.read()
    return test_org


def text_formatting(text):
    # ヘッダ，フッタ部分の除去
    text = re.split('\-{5,}', text)[2]
    text = re.split('底本：', text)[0]
    # |の除去
    text = text.replace('|', '')
    # るびの削除
    text = re.sub('《.+?》', '', text)
    # 入力注の削除
    text = re.sub('［＃.+?］', '', text)
    # 空行の削除
    text = re.sub('\n\n', '\n', text)
    text = re.sub('\r', '', text)
    return text

```

```python
url = 'https://www.aozora.gr.jp/cards/000148/files/794_ruby_4237.zip'
zip_ = '794_ruby_4237.zip'

text_org = get_text(url, zip_)


print('整形前文頭部分')
print(text_org[:600])
print()
print('*' * 100)
print()

print('整形前文末部分')
print(text_org[-300:])
print('*' * 100)
print()
```

```python
text = text_formatting(text_org)

print('整形後文頭部分')
print(text[:100])
print()
print('*' * 100)
print()

print('整形後文末部分')
print(text[-100:])
print('*' * 100)
print()
```
